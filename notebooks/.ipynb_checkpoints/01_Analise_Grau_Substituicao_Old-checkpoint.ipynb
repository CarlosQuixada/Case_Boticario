```python
dias_semana = {
    'Monday': '1-segunda-feira',
    'Tuesday': '2-terça-feira',
    'Wednesday': '3-quarta-feira',
    'Thursday': '4-quinta-feira',
    'Friday': '5-sexta-feira',
    'Saturday': '6-sábado',
    'Sunday': '7-domingo'
}

```

```python
df_rupturas = pd.read_csv('../data/processed/dados_ruptura.csv')
df_rupturas['dt_estoque'] = pd.to_datetime(df_rupturas['dt_estoque'])
# df_rupturas['dia_semana_en'] = df_rupturas['dt_estoque'].dt.day_name()
# df_rupturas['ano_mes'] = df_rupturas['dt_estoque'].dt.strftime('%Y-%m')
# df_rupturas['dia_semana_pt'] = df_rupturas['dia_semana_en'].map(dias_semana)
df_rupturas['is_ruptura'] = df_rupturas['is_ruptura'].apply(int)
df_rupturas.head()
```

```python
df_similares = pd.read_csv('../data/raw/tb_similares.csv')
df_similares.rename(columns={"cod_material": "cod_produto"}, inplace=True)
df_similares.head()
```

```python
df_estoque = pd.read_csv('../data/raw/tb_estoques.csv')
df_estoque['dt_estoque'] = pd.to_datetime(df_estoque['dt_estoque'])
df_estoque.head()
```

```python
df_vendas = pd.read_csv('../data/raw/tb_vendas.csv')
df_vendas['dt_venda'] = pd.to_datetime(df_vendas['dt_venda'])
df_vendas = df_vendas.drop_duplicates(subset=['dt_venda', 'cod_pedido', 'canal', 'des_estado_franquia', 'key_loja', 'cod_material']).reset_index(drop=True)
df_vendas.rename(columns={"cod_material": "cod_produto"}, inplace=True)
df_vendas.head()
```

```python
df_ruptura_similar = df_rupturas.merge(df_similares, on='cod_produto', how='left')
df_ruptura_similar.head()
```

```python
df_ruptura_similar = df_ruptura_similar.merge(
    df_estoque,
    left_on=['cod_material_recom', 'key_loja', 'dt_estoque'],
    right_on=['cod_produto', 'key_loja', 'dt_estoque'],
    how='left',
    suffixes=('', '_similar')
)

df_ruptura_similar.head()

# Filtrar se o similar tinha estoque disponível (> 0)
df_ruptura_similar['similar_disponivel'] = df_ruptura_similar['qt_estoque_similar'] > 0
df_ruptura_similar['similar_disponivel'] = df_ruptura_similar['similar_disponivel'].apply(int)
df_ruptura_similar.head()
```

## 1.3. Quantas dessas situações resultaram em vendas de similares? (Geral) -> Fazendo

```python
rupturas_similares_disponiveis = df_ruptura_similar[(df_ruptura_similar['is_ruptura']==1)&(df_ruptura_similar['similar_disponivel']==1)]
rupturas_similares_disponiveis.head()
```

```python
rupturas_similares_vendas = rupturas_similares_disponiveis.merge(
    df_vendas,
    left_on=['cod_material_recom', 'key_loja', 'dt_estoque'],
    right_on=['cod_produto', 'key_loja', 'dt_venda'],
    how='left',
    suffixes=('', '_venda')
)

# Criar flag: houve venda do similar?
rupturas_similares_vendas['venda_similar'] = ~rupturas_similares_vendas['cod_pedido'].isna()
rupturas_similares_vendas.head()
```

```python
# Agregar por ruptura
resultado_venda_similar = (
    rupturas_similares_vendas
    .groupby(['cod_produto', 'key_loja', 'dt_estoque'])['venda_similar']
    .max()
    .reset_index()
)

# Contar rupturas cobertas por venda de similar
rupturas_com_venda_similar = resultado_venda_similar['venda_similar'].sum()
rupturas_totais_com_similar_disponivel = resultado_venda_similar.shape[0]

# Taxa de sucesso
taxa_sucesso_venda_similar = rupturas_com_venda_similar / rupturas_totais_com_similar_disponivel

print(f"Taxa de rupturas cobertas com venda de similar: {taxa_sucesso_venda_similar:.2%}")
```

### 2.1. A substituição é maior para quais rankings de similaridade? ->

```python
similares_vendidos = rupturas_similares_vendas[rupturas_similares_vendas['venda_similar'] == True]
similares_vendidos
```

```python
# Ordenar por ranking
similares_vendidos = similares_vendidos.sort_values(['cod_produto', 'key_loja', 'dt_estoque', 'ranking'])

# Pegar o primeiro vendido por ruptura
vendido_por_ranking = (
    similares_vendidos
    .groupby(['cod_produto', 'key_loja', 'dt_estoque'])
    .first()
    .reset_index()
)

vendido_por_ranking
```

```python
ranking_substituicao = vendido_por_ranking['ranking'].value_counts().reset_index().sort_values('count')

# Calcular % de participação
total = ranking_substituicao['count'].sum()
ranking_substituicao['percentual'] = ranking_substituicao['count'] / total * 100
ranking_substituicao
```

## 2.2. O canal de venda influencia na substituição? -> REVISAR 

```python
rupturas_similares_vendas.fillna(value={'canal':'sem_venda'}, inplace=True)
rupturas_similares_vendas
```

```python
# Agregar por ruptura: se houve venda de similar
substituicoes = (
    rupturas_similares_vendas
    .groupby(['cod_produto', 'key_loja', 'dt_estoque', 'canal'])['venda_similar']
    .max()
    .reset_index()
)

# Agora calcular taxa por canal
taxa_por_canal = (
    substituicoes
    .groupby('canal')['venda_similar']
    .agg(['count', 'sum'])  # count = rupturas com similar disponível, sum = substituições feitas
    .reset_index()
    .rename(columns={'count': 'rupturas_com_similar', 'sum': 'substituicoes'})
)

total_ruptura = df_rupturas[df_rupturas['is_ruptura']==1].shape[0]
# Calcular taxa
taxa_por_canal['taxa_substituicao'] = taxa_por_canal['substituicoes'] / total_ruptura * 100
taxa_por_canal
```

## 2.3. Há variação por categoria, subcategoria ou linha do produto? ->

- Alguns produtos paresentam mais de uma categoria na base de vendas, com isso para evitar duplicatas vou filtrar subcategoria e linha de produto

- Em um cenário real o ideal seria conferir com os responsáveis por esses dados qual categoria devo considerar e assim utilizar a informação

```python
df_desc_prod = df_vendas[['cod_produto', 'sub_categoria_produto', 'linha_produto']].drop_duplicates()
```

```python
df_desc_prod['cod_produto'].value_counts()
```

```python
df_rup_sim_disp = df_ruptura_similar[(df_ruptura_similar['is_ruptura']==1)&(df_ruptura_similar['similar_disponivel']==1)]
df_rup_sim_disp.shape
```

```python
df_rup_sim_disp = df_rup_sim_disp.merge(df_desc_prod, on=['cod_produto'],how='left')
df_rup_sim_disp.head()
```

```python
df_rup_sim_disp.shape
```

```python
rupturas_similares_vendas = df_rup_sim_disp.merge(
    df_vendas,
    left_on=['cod_produto_similar', 'key_loja', 'dt_estoque'],
    right_on=['cod_produto', 'key_loja', 'dt_venda'],
    how='left',
    suffixes=('', '_venda')
)

# Criar flag: houve venda do similar?
rupturas_similares_vendas['venda_similar'] = ~rupturas_similares_vendas['cod_pedido'].isna()
rupturas_similares_vendas.head()
```

```python
# Obter 1 linha por ruptura com info de produto e flag de venda
rupturas_agrupadas = (
    rupturas_similares_vendas
    .groupby(['cod_produto', 'key_loja', 'dt_estoque', 'sub_categoria_produto', 'linha_produto'])['venda_similar']
    .max()
    .reset_index()
)
rupturas_agrupadas
```

```python
taxa_por_sub_categoria = (
    rupturas_agrupadas
    .groupby('sub_categoria_produto')['venda_similar']
    .agg(['count', 'sum'])
    .reset_index()
    .rename(columns={'count': 'rupturas_com_similar', 'sum': 'substituicoes'})
)
taxa_por_sub_categoria['taxa_substituicao'] = taxa_por_sub_categoria['substituicoes'] / taxa_por_sub_categoria['rupturas_com_similar'] * 100
taxa_por_sub_categoria
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Usar base agrupada com 1 linha por ruptura
# com flag venda_similar = True ou False
# e coluna categoria_produto

# Criar tabela de contingência
contingencia = pd.crosstab(rupturas_agrupadas['sub_categoria_produto'], rupturas_agrupadas['venda_similar'])

chi2, p, dof, expected = chi2_contingency(contingencia)

print(f"Estatística Qui-quadrado: {chi2:.2f}")
print(f"p-valor: {p:.4f}")

if p < 0.05:
    print("Conclusão: Há diferença estatisticamente significativa entre as sub_categoria_produto quanto à taxa de substituição.")
else:
    print("Conclusão: Não há diferença estatística significativa entre as sub_categoria_produto.")
```

```python
taxa_por_linha_produto = (
    rupturas_agrupadas
    .groupby('linha_produto')['venda_similar']
    .agg(['count', 'sum'])
    .reset_index()
    .rename(columns={'count': 'rupturas_com_similar', 'sum': 'substituicoes'})
)
taxa_por_linha_produto['taxa_substituicao'] = taxa_por_linha_produto['substituicoes'] / taxa_por_linha_produto['rupturas_com_similar'] * 100
taxa_por_linha_produto
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Usar base agrupada com 1 linha por ruptura
# com flag venda_similar = True ou False
# e coluna categoria_produto

# Criar tabela de contingência
contingencia = pd.crosstab(rupturas_agrupadas['linha_produto'], rupturas_agrupadas['venda_similar'])

chi2, p, dof, expected = chi2_contingency(contingencia)

print(f"Estatística Qui-quadrado: {chi2:.2f}")
print(f"p-valor: {p:.4f}")

if p < 0.05:
    print("Conclusão: Há diferença estatisticamente significativa entre as linha_produto quanto à taxa de substituição.")
else:
    print("Conclusão: Não há diferença estatística significativa entre as linha_produto.")
```
